## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import json, requests
url = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(url.text)
print('共有 ', len(data), ' 筆資料。\n每一筆資料包含:')
for i in data[0]:
    print(i)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for i in data:
    print('標題: ', i['title'], ' \n貼文時間: ', i['createdAt'], ' \n留言人數: ', i['commentCount'], ' \n按讚人數: ', i['likeCount'], '\n')
```

```python
# 3. 計算熱門/最新文章的「平均留言人數」與「平均按讚人數」
truecommend, truelike, newcommend, newlike = 0
#truelike = 0
#newcommend = 0
#newlike = 0
for i in data:
    truecommend = truecommend + i['commentCount']
    truelike = truelike + i['likeCount']
newurl = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
newdata = json.loads(newurl.text)
for i in newdata:
    newcommend = newcommend + i['commentCount']
    newlike = newlike + i['likeCount']
print('熱門文章:\n平均留言人數: ', truecommend/len(data), '\n平均按讚人數: ', truelike/len(data), '\n\n最新文章:\n平均留言人數: ', newcommend/len(newdata), '\n平均按讚人數: ', newlike/len(newdata))
```
